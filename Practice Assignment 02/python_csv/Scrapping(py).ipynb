```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Website

```python
URL = "https://www.sim.edu.sg/"
page = requests.get(URL)
page.text
```

### Webscrapping

#### Getting text of html

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

#### Locate Contact Information

```python
footer = soup.find('footer')
spans = footer.find_all('span')
spans
```

#### Cleaning Retrieved Contact Information

```python
index = 0
email_lst = []
phone_lst = []
address_lst = []

for span in spans:
    if ("Email" in span.text):
        email = spans[index+1].text
        email = email.strip()
        email_lst.append(email)

    if("Phone" in span.text):
        phone = spans[index+1].text

        # 8 Number + 1 Space = 9 Characters
        phone = phone [:10]
        phone = phone.strip()
        phone_lst.append(phone)

    if("Address" in span.text):
        address = spans[index+1].text
        address_end = address.find("\n")
        address = address[:address_end]
        address = address.strip()
        address_lst.append(address)

    index += 1
```

```python
print(email_lst)
phone_lst.append("") ## To deal with the different length
print(phone_lst)
print(address_lst)
```

#### Create Contact Dictionary

```python
contact_dict = {'Email': email_lst, 'Phone': phone_lst, 'Address': address_lst}
contact_dict
```

#### Create Data Frame

```python
contact_df = pd.DataFrame(contact_dict)
contact_df
```

#### Save to CSV File

```python
contact_df.to_csv("contact.csv", index=False, header=True)
```
